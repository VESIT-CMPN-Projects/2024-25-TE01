### Unsloth

````python
from unsloth import FastLanguageModel
import torch
max_seq_length = 2048 # Choose any! We auto support RoPE Scaling internally!
dtype = None # None for auto detection. Float16 for Tesla T4, V100, Bfloat16 for Ampere+
load_in_4bit = True # Use 4bit quantization to reduce memory usage. Can be False.

# 4bit pre quantized models we support for 4x faster downloading + no OOMs.
fourbit_models = [
    "unsloth/mistral-7b-bnb-4bit",
    "unsloth/mistral-7b-instruct-v0.2-bnb-4bit",
    "unsloth/llama-2-7b-bnb-4bit",
    "unsloth/llama-2-13b-bnb-4bit",
    "unsloth/codellama-34b-bnb-4bit",
    "unsloth/tinyllama-bnb-4bit",
    "unsloth/gemma-7b-bnb-4bit", # New Google 6 trillion tokens model 2.5x faster!
    "unsloth/gemma-2b-bnb-4bit",
] # More models at https://huggingface.co/unsloth

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = "unsloth/mistral-7b-instruct-v0.3-bnb-4bit", # Choose ANY! eg teknium/OpenHermes-2.5-Mistral-7B
    max_seq_length = max_seq_length,
    dtype = dtype,
    load_in_4bit = load_in_4bit,
    # token = "hf_...", # use one if using gated models like meta-llama/Llama-2-7b-hf
)
````

We now add LoRA adapters so we only need to update 1 to 10% of all parameters!

````python
model = FastLanguageModel.get_peft_model(
    model,
    r = 8,  # Lower rank is usually sufficient for small datasets
    target_modules = ["q_proj", "k_proj", "v_proj", "o_proj",
                      "gate_proj", "up_proj", "down_proj"],
    lora_alpha = 16,
    lora_dropout = 0.05,  # Small dropout for regularization
    bias = "none",
    use_gradient_checkpointing = "unsloth",  # Helps with memory
    random_state = 42,
    use_rslora = False,
    loftq_config = None,
)
````

<a name="Data"></a>
### Data Prep
We now use the `ChatML` format for conversation style finetunes. We use [Open Assistant conversations](https://huggingface.co/datasets/philschmid/guanaco-sharegpt-style) in ShareGPT style. ChatML renders multi turn conversations like below:

```
<|im_start|>system
You are a helpful assistant.<|im_end|>
<|im_start|>user
What's the capital of France?<|im_end|>
<|im_start|>assistant
Paris.
```

**[NOTE]** To train only on completions (ignoring the user's input) read TRL's docs [here](https://huggingface.co/docs/trl/sft_trainer#train-on-completions-only).

We use our `get_chat_template` function to get the correct chat template. We support `zephyr, chatml, mistral, llama, alpaca, vicuna, vicuna_old` and our own optimized `unsloth` template.

Normally one has to train `<|im_start|>` and `<|im_end|>`. We instead map `<|im_end|>` to be the EOS token, and leave `<|im_start|>` as is. This requires no additional training of additional tokens.

Note ShareGPT uses `{"from": "human", "value" : "Hi"}` and not `{"role": "user", "content" : "Hi"}`, so we use `mapping` to map it.

For text completions like novel writing, try this [notebook](https://colab.research.google.com/github/unslothai/notebooks/blob/main/nb/Mistral_(7B)-Text_Completion.ipynb).

````python
import pandas as pd

df = pd.read_csv("/content/remaining_480_rows.csv")
print(df.columns.tolist())
````

````python
from datasets import Dataset
import pandas as pd

# Load dataset
df = pd.read_csv("/content/remaining_480_rows.csv")

# Convert to Hugging Face Dataset
dataset = Dataset.from_pandas(df)

# Format each example into prompt + response text
def formatting_prompts_func(example):
    prompt = (
        f"Classify this network activity:\n"
        f"Source IP: {example['Source IP']}\n"
        f"Destination IP: {example['Destination IP']}\n"
        f"Source Port: {example['Source Port']}\n"
        f"Destination Port: {example['Destination Port']}\n"
        f"Protocol: {example['Protocol']}\n"
        f"Flow Duration: {example['Flow Duration']}\n"
        f"Total Fwd Packets: {example['Total Fwd Packets']}\n"
        f"Total Backward Packets: {example['Total Backward Packets']}\n"
        f"Flow Bytes/s: {example['Flow Bytes/s']}\n"
        f"Label:"
    )
    response = example["Label"]
    return { "text": f"<s>[INST] {prompt} [/INST] {response}</s>" }

# Apply formatting
dataset = dataset.map(formatting_prompts_func)
````

Let's see how the `ChatML` format works by printing the 5th element

````python
dataset[5]["text"]
````

If you're looking to make your own chat template, that also is possible! You must use the Jinja templating regime. We provide our own stripped down version of the `Unsloth template` which we find to be more efficient, and leverages ChatML, Zephyr and Alpaca styles.

More info on chat templates on [our wiki page!](https://github.com/unslothai/unsloth/wiki#chat-templates)

````python
unsloth_template = \
    "{{ bos_token }}"\
    "{{ 'You are a helpful assistant to the user\n' }}"\
    "{% for message in messages %}"\
        "{% if message['role'] == 'user' %}"\
            "{{ '>>> User: ' + message['content'] + '\n' }}"\
        "{% elif message['role'] == 'assistant' %}"\
            "{{ '>>> Assistant: ' + message['content'] + eos_token + '\n' }}"\
        "{% endif %}"\
    "{% endfor %}"\
    "{% if add_generation_prompt %}"\
        "{{ '>>> Assistant: ' }}"\
    "{% endif %}"
unsloth_eos_token = "eos_token"

# Update tokenizer with new template mapping
if False:
    tokenizer = get_chat_template(
        tokenizer,
        chat_template = (unsloth_template, unsloth_eos_token,),  # Provide template and EOS token
        mapping = {
            "role": "from",
            "content": "value",
            "user": "human",
            "assistant": "gpt"
        },  # Map roles and content based on the ShareGPT style
        map_eos_token = True,  # Ensure EOS token is mapped correctly (e.g., <|im_end|> to </s>)
    )
````

<a name="Train"></a>
### Train the model
Now let's use Huggingface TRL's `SFTTrainer`! More docs here: [TRL SFT docs](https://huggingface.co/docs/trl/sft_trainer). We do 60 steps to speed things up, but you can set `num_train_epochs=1` for a full run, and turn off `max_steps=None`. We also support TRL's `DPOTrainer`!

````python
from trl import SFTTrainer
from transformers import TrainingArguments
from unsloth import is_bfloat16_supported

# Set up the Trainer
trainer = SFTTrainer(
    model = model,  # Use your already loaded model
    tokenizer = tokenizer,  # Use your tokenizer
    train_dataset = dataset,  # Your dataset that has been formatted
    dataset_text_field = "text",  # Field in the dataset holding text
    max_seq_length = max_seq_length,  # Max sequence length for the model
    dataset_num_proc = 2,  # Number of processes to use for loading dataset
    packing = False,  # Disable packing to avoid issues with long sequences
    args = TrainingArguments(
        per_device_train_batch_size = 2,  # Batch size per device
        gradient_accumulation_steps = 4,  # Gradient accumulation steps for large models
        warmup_steps = 5,  # Warmup steps for learning rate scheduler
        max_steps = 60,  # Total number of steps for training
        learning_rate = 2e-4,  # Learning rate for training
        fp16 = not is_bfloat16_supported(),  # Use fp16 if bfloat16 is not supported
        bf16 = is_bfloat16_supported(),  # Use bf16 if supported
        logging_steps = 1,  # Log every step
        optim = "adamw_8bit",  # Optimizer to use
        weight_decay = 0.01,  # Weight decay regularization
        lr_scheduler_type = "linear",  # Linear learning rate scheduler
        seed = 3407,  # Set random seed for reproducibility
        output_dir = "outputs",  # Directory to save outputs
        report_to = "none",  # Disables logging to external tools (WandB, etc)
    ),
)

# Start training
trainer.train()
````

````python
# @title Show current memory stats
gpu_stats = torch.cuda.get_device_properties(0)
start_gpu_memory = round(torch.cuda.max_memory_reserved() / 1024 / 1024 / 1024, 3)
max_memory = round(gpu_stats.total_memory / 1024 / 1024 / 1024, 3)
print(f"GPU = {gpu_stats.name}. Max memory = {max_memory} GB.")
print(f"{start_gpu_memory} GB of memory reserved.")
````

<a name="Inference"></a>
### Inference
Let's run the model! Since we're using `ChatML`, use `apply_chat_template` with `add_generation_prompt` set to `True` for inference.

````python
test_data = pd.read_csv('/content/sample_20_rows.csv')
X_test = test_data.drop('Label', axis=1)
y_test = test_data['Label']
````

````python
from unsloth.chat_templates import get_chat_template

system_message = """
You are a cybersecurity expert trained to analyze network flow logs and determine whether the traffic is benign or part of a Distributed Denial of Service (DDoS) attack.

Use the ReAct (Reasoning + Acting) approach:
1. First, REASON through the meaning of the network features such as IPs, ports, protocols, flow duration, packet counts, and byte rates.
2. Then, ACT by concluding whether the flow is BENIGN or DDOS based on patterns, anomalies, or suspicious behavior.

Be concise but clear in your reasoning, and always end with the final classification in the format: "Final Answer: BENIGN" or "Final Answer: DDOS".
"""

tokenizer = get_chat_template(
    tokenizer,
    chat_template = "chatml", # Supports zephyr, chatml, mistral, llama, alpaca, vicuna, vicuna_old, unsloth
    mapping = {"role": "from", "content": "value","role" : "from", "content" : "value", "user" : "human", "assistant" : "gpt"}, # ShareGPT style
    map_eos_token = True, # Maps <|im_end|> to </s> instead
)

FastLanguageModel.for_inference(model)

predictions = []
for i in range(20):
    messages = [
        {"from": "system", "value": system_message},
        {"from": "human", "value": f"""
- Source IP: {X_test['Source IP'][i]}
- Destination IP: {X_test['Destination IP'][i]}
- Source Port: {X_test['Source Port'][i]}
- Destination Port: {X_test['Destination Port'][i]}
- Protocol: {X_test['Protocol'][i]}
- Flow Duration: {X_test['Flow Duration'][i]}
- Total Fwd Packets: {X_test['Total Fwd Packets'][i]}
- Total Backward Packets: {X_test['Total Backward Packets'][i]}
- Flow Bytes/s: {X_test['Flow Bytes/s'][i]}
        """},
    ]
    inputs = tokenizer.apply_chat_template(
        messages,
        tokenize=True,
        add_generation_prompt=True,
        return_tensors="pt",
    ).to("cuda")

    outputs = model.generate(input_ids=inputs, max_new_tokens=1024, use_cache=True, temperature=0.8, min_p=0.1)
    decoded_output = tokenizer.batch_decode(outputs)[0]
    print(f"Response for index {i}:\n")
    print(decoded_output)

    # Optionally store predictions for later evaluation
    predictions.append(decoded_output)
````

 You can also use a `TextStreamer` for continuous inference - so you can see the generation token by token, instead of waiting the whole time!

### GGUF / llama.cpp Conversion
To save to `GGUF` / `llama.cpp`, we support it natively now! We clone `llama.cpp` and we default save it to `q8_0`. We allow all methods like `q4_k_m`. Use `save_pretrained_gguf` for local saving and `push_to_hub_gguf` for uploading to HF.

Some supported quant methods (full list on our [Wiki page](https://github.com/unslothai/unsloth/wiki#gguf-quantization-options)):
* `q8_0` - Fast conversion. High resource use, but generally acceptable.
* `q4_k_m` - Recommended. Uses Q6_K for half of the attention.wv and feed_forward.w2 tensors, else Q4_K.
* `q5_k_m` - Recommended. Uses Q6_K for half of the attention.wv and feed_forward.w2 tensors, else Q5_K.

````python
# Save to 8bit Q8_0
if False: model.save_pretrained_gguf("model", tokenizer,)
if False: model.push_to_hub_gguf("hf/model", tokenizer, token = "")

# Save to 16bit GGUF
if False: model.save_pretrained_gguf("model", tokenizer, quantization_method = "f16")
if False: model.push_to_hub_gguf("hf/model", tokenizer, quantization_method = "f16", token = "")

# Save to q4_k_m GGUF
if False: model.save_pretrained_gguf("model", tokenizer, quantization_method = "q4_k_m")
if False: model.push_to_hub_gguf("hf/model", tokenizer, quantization_method = "q4_k_m", token = "")
````
